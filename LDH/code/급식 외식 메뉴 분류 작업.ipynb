```python
def zacard(A,B):
    intersection_cardinality = len(set.intersection(*[set(A), set(B)]))
    union_cardinality = len(set.union(*[set(A), set(B)]))
    similar = intersection_cardinality / float(union_cardinality)
    
    return similar


#ex) 새송이 들깨볶음 = 총알새송이볶음 위해
```

# dict만들고 돌면서 분류하기

```python
def check(menu, tmp_dict, eps):
    value=0
    tmp_max=0
    key=0
    
    for i in tmp_dict.keys():
        if zacard(menu, i) >=eps and zacard(menu,i) > tmp_max:
            value=1
            key = i
    
    return value, key
```

```python
df_category = df_small['상품범주(세)'].unique()
```

```python
cat_idx = []
cat_dict = {}
i=0

for cat in df_category:
    
    tmp_set = df_small[df_small['상품범주(세)']==cat]
    tmp_set_1 = tmp_set['상품명1']
    tmp_dict = {}
    i+=1
    for menu in tmp_set_1:
        
        if len(tmp_dict)==0:
            tmp_dict[menu] = i
            cat_idx.append(i)
            cat_dict[i] = menu
            
            continue
        
        check_value,key = check(menu,tmp_dict,0.4)
        
        if check_value ==1:#있으면
            cat_idx.append(tmp_dict[key])
        else:
            i +=1
            tmp_dict[menu] = i
            cat_idx.append(i)
            cat_dict[i] = menu
            
        
            
    
            
            
```
