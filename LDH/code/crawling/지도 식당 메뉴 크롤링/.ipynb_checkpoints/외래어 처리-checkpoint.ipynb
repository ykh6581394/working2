```python
def word_proc(menu_nm):

    en_nm = translator.translate(menu_nm,src="ko",dest="en").text

    result = translator.translate(en_nm,src="en",dest="ko").text
    
    return result
```

```python
translator.translate('사이다',src="ko",dest="en").text
```

```python
import pandas as pd
dt = pd.DataFrame({'메뉴':dt})
```

```python
changed_menu = dt['메뉴'].apply(word_proc)
```
