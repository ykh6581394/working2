```python
def jacard(set1, set2):
    set1= set(set1)
    set2=set(set2)
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return intersection / union  
```

```python
#특수문자들 index 전부 구해서 그 사이에 있는 단어들 split 해서 [0]만 뽑아냄
def make_ingred(dt):
    eps=0.5
    
    print(dt)
    idx = dt.find(']')
    dt = dt[idx+1:]
    
    tmp_ingred = dt.split()
    #print(tmp_ingred)
    tmp_ingred1 = []

    #tmp_ingred 돌면서 특수문자 없애기
    for i in tmp_ingred:
        regex = r"[\(\)\[\]\|]"  # (, ), [, ], | 특수문자를 포함한 정규식 패턴
        i = re.sub(regex, "", i)
        if i!= "":
            tmp_ingred1.append(i)
        
    tmp_ingred = tmp_ingred1

    ingred_list=[]

    for i in tmp_ingred:
        
        score=0
        tmp_score=0
        #print('1')
        #print(i+"2")
        if ord('가') <= ord(i[0]) <= ord('힣') and ord('가') <= ord(i[-1]) <= ord('힣'):
            
            for menu in sku_dict['Dictionary_re'].unique():
                ## 자카드로 변경
               
                tmp_score = jacard(i, menu)
                #print(tmp_score)
                
                if tmp_score>score:
                    score= tmp_score
            #print(i)
            #print(score)
            if score > eps:
                ingred_list.append(i)
            
    return ingred_list
```

```python
ingred_col = df['재료'].apply(make_ingred)
recipe = pd.concat([df['음식'], ingred_col], axis=1)
```
