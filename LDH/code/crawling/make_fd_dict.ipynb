# 음식 사전을 만들기

# 음식 레시피 크롤링해서 음식 이름 & 재료 전부 넣기
# but 이런데는 소스는 없으니 소스 들어간건 원 데이터에서 넣기

```python
import pandas as pd
import numpy as np
from urllib.request import urlopen
from bs4 import BeautifulSoup



result = set()

for nation in [10,20,30,40,50,60,80,90]:
    page_number=1
    print(nation)
    while True:
        print(page_number)

        #print(page_number)
        html = urlopen("https://www.menupan.com/Cook/recipere.asp?nation="+str(nation)+"&page="+str(page_number))
        bsObject = BeautifulSoup(html, "lxml", from_encoding='utf-8')

        #메뉴 수집
        menus = bsObject.findAll('span', {"class":"link"})
        
        #print(menus)
        if len(menus)==0:
            break
    
        for menu in menus:
            menu_tmp =menu.get_text()
            result.add(menu_tmp)
    
        # 재료 수집
        ingredients= bsObject.findAll('td', {"valign":"top"})
        for ingredient in ingredients[5:]:
            tmp = ingredient.findAll('td')
            for i in tmp:
                tmp_text = i.get_text()
                if tmp_text ==None:
                    continue
                else:
                    if '주재료:' in tmp_text:
                        tmp_ingred = tmp_text[4:].replace(',','').split()
                        for ing in tmp_ingred:
                            result.add(ing)
        page_number +=1

        #print(list(result)[:3])
```

```python
import pandas as pd
result = list(result)
fd_dict = pd.DataFrame({"사전":result})
```
