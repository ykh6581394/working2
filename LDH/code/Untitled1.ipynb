```python
def jacard(set1, set2):
    set1= set(set1)
    set2=set(set2)
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return intersection / union   

def transform(ingred,sku_dict):
    #상품명중 유사도 가장 높은거 key로 가져와서 dictionary_re 로 return하기
    score=0
    for i in range(len(sku_dict)):
        tmp_score = jacard(ingred, sku_dict['상품명'][i])
        
        if tmp_score>score:
            score= tmp_score
            result = sku_dict['Dictionary_re'][i]
            
        if tmp_score==1:
            result= sku_dict['Dictionary_re'][i]
            return result
        
        if score<=0.2:
            return "없음"
        
    return result

def transform2(ingred, cols):
    #print(ingred)
    check=0
    for col in cols:
        if ingred in col:
           check = col
    return check   
```
